```python
import pandas as pd
import glob
```

### Dim Table

```python
# List all Parquet files in folder
files_dim = glob.glob("dim_table/*.parquet")
files_dim
```

```python
# Read and concatenate them
df_dim = pd.concat([pd.read_parquet(f, engine="fastparquet") for f in files_dim], ignore_index=True)
df_dim.head()
```

```python
# Check basic info
df_dim.info()
```

```python
df_dim[['PROVIDER_NUM', 'AVERAGE_NUMBER_OF_RESIDENTS_PER_DAY', 'REPORTED_TOTAL_NURSE_STAFFING_HOURS_PER_RESIDENT_PER_DAY', 'TOTAL_NUMBER_OF_NURSE_STAFF_HOURS_PER_RESIDENT_PER_DAY_ON_THE_WEEKEND']].head(10)
```
